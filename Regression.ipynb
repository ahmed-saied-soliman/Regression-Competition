```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn import svm
from sklearn import linear_model
```

## SGD Implementation

```python
data=pd.read_csv("C:/Users/Ahmed Saied/Downloads/Compressed/knn/reg_train.csv")
x_train=data.iloc[:,2:14]
y_train=data.iloc[:,16:]
training= linear_model.SGDRegressor()
training.fit(x_train,y_train)
data_test=pd.read_csv("C:/Users/Ahmed Saied/Downloads/Compressed/knn/reg_test.csv")
x_test=data_test.iloc[:,2:14]
x_test.info()
y_test=training.predict(x_test)
print(pd.DataFrame(y_test))
result=pd.DataFrame(y_test)
result.to_csv("C:/Users/Ahmed Saied/Downloads/Compressed/knn/R_result_sgd.csv")
```

## SVM Implementation

```python
data=pd.read_csv("C:/Users/Ahmed Saied/Downloads/Compressed/knn/reg_train.csv")
x_train=data.iloc[:,2:14]
y_train=data.iloc[:,16:]
training=svm.SVR()
training.fit(x_train,y_train.values.ravel())
data_test=pd.read_csv("C:/Users/Ahmed Saied/Downloads/Compressed/knn/reg_test.csv")
x_test=data_test.iloc[:,2:]
y_test=training.predict(x_test)
print(pd.DataFrame(y_test))
result=pd.DataFrame(y_test)
result.to_csv("C:/Users/Ahmed Saied/Downloads/Compressed/knn/R_result_svm.csv")
```
